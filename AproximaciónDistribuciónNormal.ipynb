```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd

def f():
  u = rd.uniform(0,1)
  x = (u**0.135 - (1-u)**0.135)/0.1975
  return x

def experimento(n):
  lista = []
  for i in range(n):
    lista.append(f())
  return lista

prueba = experimento(100000)
print(np.mean(prueba))
print(np.std(prueba))
plt.hist(prueba, bins=100)
plt.show()
```
