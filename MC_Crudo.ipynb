```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir la función
def f(x):
    return np.sqrt(np.arctan(x))
```

```python
N = 10000
x_random = np.random.uniform(0, 1, N)

print(x_random)
```

```python
# Evaluar la función en estos puntos
f_values = f(x_random)

sum_f_values = np.sum(f_values)
promedio = sum_f_values / N

print(promedio)
```

```python
# Graficar la función
x = np.linspace(0, 1, 500)
y = f(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$f(x) = \sqrt{\arctan(x)}$', color='b')

plt.axhline(promedio, color='black', linestyle='--', label=f'Promedio ≈ {promedio:.4f}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Estimación del valor promedio de $\\sqrt{\\arctan(x)}$')
plt.show()

print(f'Suma de valores: {sum_f_values:.4f}')
print(f'Valor promedio estimado: {promedio:.4f}')
```
