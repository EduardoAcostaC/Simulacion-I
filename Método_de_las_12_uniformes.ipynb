##  Método de Suma de Uniformes y Distribución Triangular

### Método de Suma de Uniformes

Este método se basa en el **Teorema Central del Límite**, el cual dice que la suma de varias variables aleatorias independientes tiende a una distribución normal.

Pasos:

1. Generar 12 números uniformes entre 0 y 1.
2. Sumar y restar 6 para centrar la media en 0:  
   $$
   X = \sum_{i=1}^{12} U_i - 6
   $$
3. Escalar con media $$ \mu $$ y desviación estándar $$ \sigma $$:  
   $$
   Z = \mu + \sigma \cdot X
   $$

Este método aproxima una **distribución normal** usando solo números uniformes.

---

### Distribución Triangular

Se define por tres parámetros: mínimo \( a \), máximo \( b \), y moda \( c \). Es útil cuando se conocen límites y un valor más probable, pero se tiene poca información.

La función de densidad tiene forma triangular y puede ser simétrica o no, según los valores de \( a, b, c \).

---

### Comparación Rápida

| Característica     | Suma de Uniformes     | Distribución Triangular |
|--------------------|-----------------------|--------------------------|
| Forma              | Aproximadamente normal| Triangular               |
| Parámetros         | $$ \mu, \sigma $$    | \( a, b, c \)            |
| Uso                | Simular normales      | Modelar incertidumbre    |


```python

import random as rd
import numpy as np
import matplotlib.pyplot as plt

#Variables
mu = 2
sigma = 3
lista = []

#Funciones
def sumatoria():
    sumatoria = 0
    for _ in range(12):
        num = rd.uniform(0, 1)
        sumatoria += num
    return sumatoria - 6

def experimento(n):
    for _ in range(n):
        x = mu + sigma * sumatoria()
        lista.append(x)
    return lista

#Experimento
n = 100000
resultados = experimento(n)

#Valores experimentales
media_exp = np.mean(resultados)
desv_est_exp = np.std(resultados)

#Resultados
print(f"Media teórica: {mu}")
print(f"Media experimental: {media_exp:.4f}")
print(f"Diferencia: {abs(mu - media_exp):.4f}")

print(f"Desviación estándar teórica: {sigma}")
print(f"Desviación estándar experimental: {desv_est_exp:.4f}")
print(f"Diferencia: {abs(sigma - desv_est_exp):.4f}")

plt.hist(resultados, bins=50, color='skyblue', edgecolor='black')
plt.title("Histograma de los resultados del experimento")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```
